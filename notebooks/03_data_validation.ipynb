- **Uniqueness**ï¼šéƒ½é“åºœçœŒã‚­ãƒ¼ã®ä¸€æ„æ€§ï¼ˆé‡è¤‡ãƒ»æ¬ è½ï¼‰
- **Sanity**ï¼šãƒ¬ãƒ³ã‚¸ãƒ»åˆ†å¸ƒã®å¦¥å½“æ€§ï¼ˆæ˜ã‚‰ã‹ãªç•°å¸¸å€¤ï¼‰
- **Categorical Inspection**ï¼šã‚«ãƒ†ã‚´ãƒªå€¤ã®ã‚µãƒ³ãƒ—ãƒ«æŠ½å‡ºï¼ˆè¡¨è¨˜æºã‚Œï¼æ··å…¥æ¤œçŸ¥ï¼‰

## âš  Notes

- æœ¬Notebookã¯ã€Œé›†è¨ˆãƒ‡ãƒ¼ã‚¿ã®å“è³ªç¢ºèªã€ã§ã‚ã‚Šã€å€‹äººãƒ‡ãƒ¼ã‚¿ã®å› æœæ¨è«–ã‚’è¡Œã†ã‚‚ã®ã§ã¯ãªã„  
- ç•°å¸¸å€¤ã‚„æ¬ æãŒè¦‹ã¤ã‹ã£ãŸå ´åˆã€**ä¿®æ­£æ–¹é‡ï¼ˆé™¤å¤–ï¼è£œå®Œï¼å†å–å¾—ï¼‰** ã‚’æ˜ç¤ºã—ã€ä¸‹æµå·¥ç¨‹ã®å…¥åŠ›ã‚’å®‰å®šåŒ–ã•ã›ã‚‹


```python
import pandas as pd  # ãƒ‡ãƒ¼ã‚¿åˆ†æã®ä¸»å½¹ãƒ„ãƒ¼ãƒ«ã€Œpandasã€ã‚’pdã¨ã„ã†åå‰ã§ä½¿ãˆã‚‹ã‚ˆã†ã«ã™ã‚‹
import numpy as np   # æ•°å€¤è¨ˆç®—ãŒå¾—æ„ãªã€ŒNumPyã€ã‚’npã¨ã„ã†åå‰ã§æº–å‚™ï¼ˆåˆ†æã®åŸºæœ¬ã‚»ãƒƒãƒˆï¼‰
from pathlib import Path  # ãƒ•ã‚¡ã‚¤ãƒ«ã®å ´æ‰€ï¼ˆãƒ‘ã‚¹ï¼‰ã‚’ç›´æ„Ÿçš„ã«æ‰±ã†ãŸã‚ã®é“å…·ã‚’èª­ã¿è¾¼ã‚€

DATA_PATH = Path("../data/processed/nurse_data_clean.csv")  # ãƒ‡ãƒ¼ã‚¿ã®å ´æ‰€ã‚’æŒ‡å®šï¼ˆ..ã¯ã€Œ1ã¤ä¸Šã®ãƒ•ã‚©ãƒ«ãƒ€ã«æˆ»ã‚‹ã€ã¨ã„ã†æ„å‘³ï¼‰
df = pd.read_csv(DATA_PATH)  # æŒ‡å®šã—ãŸCSVãƒ•ã‚¡ã‚¤ãƒ«ã‚’èª­ã¿è¾¼ã‚“ã§ã€è¡¨ãƒ‡ãƒ¼ã‚¿ã¨ã—ã¦dfã«å…¥ã‚Œã‚‹

pd.set_option("display.max_columns", 100)  # æ¨ªã«é•·ã„ãƒ‡ãƒ¼ã‚¿ã§ã‚‚ã€å‹æ‰‹ã«çœç•¥ã›ãš200åˆ—ã¾ã§å…¨éƒ¨è¦‹ã›ã‚‹è¨­å®š
pd.set_option("display.max_rows", 100)     # ç¸¦ã«é•·ã„ãƒ‡ãƒ¼ã‚¿ã§ã‚‚ã€å‹æ‰‹ã«çœç•¥ã›ãš200è¡Œã¾ã§å…¨éƒ¨è¦‹ã›ã‚‹è¨­å®š
```

```python
# è¨­å®šã—ãŸãƒ•ã‚¡ã‚¤ãƒ«ãƒ‘ã‚¹ã‚’ãƒ­ã‚°ã«å‡ºåŠ›ã—ã¦ã€æŒ‡å®šãƒŸã‚¹ãŒãªã„ã‹ç¢ºèª
print("file:", DATA_PATH)

# ãƒ•ã‚¡ã‚¤ãƒ«ã®å®Ÿåœ¨ç¢ºèªï¼ˆTrueãªã‚‰OKã€Falseãªã‚‰ãƒ•ã‚¡ã‚¤ãƒ«ãŒç„¡ã„ï¼‰
print("exists:", DATA_PATH.exists())

# ãƒ‡ãƒ¼ã‚¿ã®æ§‹é€ ï¼ˆåˆ—åã‚„å€¤ã®å…¥ã‚Šæ–¹ï¼‰ã‚’æŠŠæ¡ã™ã‚‹ãŸã‚ã€å…ˆé ­5è¡Œã‚’ãƒ—ãƒ¬ãƒ“ãƒ¥ãƒ¼
df.head()
```

```python
# ãƒ‡ãƒ¼ã‚¿ã®è¦æ¨¡æ„Ÿï¼ˆãƒ¬ã‚³ãƒ¼ãƒ‰æ•°Ã—ã‚«ãƒ©ãƒ æ•°ã®æ¬¡å…ƒï¼‰ã‚’æŠŠæ¡
print("shape:", df.shape)

# ã‚«ãƒ©ãƒ æ•°ï¼ˆç‰¹å¾´é‡ã®ç·æ•°ï¼‰ã‚’ã‚«ã‚¦ãƒ³ãƒˆã—ã¦ç¢ºèª
print("columns:", len(df.columns))

# ãƒ‡ãƒ¼ã‚¿å‹ï¼ˆDtypeï¼‰ã€æ¬ æå€¤ï¼ˆNullï¼‰ã®æœ‰ç„¡ã€ãƒ¡ãƒ¢ãƒªä½¿ç”¨é‡ã‚’ä¸€æ‹¬ã§ãƒ—ãƒ­ãƒ•ã‚¡ã‚¤ãƒªãƒ³ã‚°
df.info()
```

```python
# å…¨ã‚«ãƒ©ãƒ åï¼ˆã‚¹ã‚­ãƒ¼ãƒï¼‰ã‚’ã€ãƒãƒ³ãƒ‰ãƒªãƒ³ã‚°ã—ã‚„ã™ã„æ¨™æº–ãƒªã‚¹ãƒˆå½¢å¼ã§å–å¾—
df.columns.tolist()
```

```python
# ã‚«ãƒ©ãƒ åï¼ˆã‚¹ã‚­ãƒ¼ãƒï¼‰ã®é‡è¤‡ã‚’åˆ¤å®šã—ã€è©²å½“ã™ã‚‹åˆ—åã ã‘ã‚’æ¨™æº–ãƒªã‚¹ãƒˆå½¢å¼ã§æŠ½å‡º
dup_cols = df.columns[df.columns.duplicated()].tolist()

# é‡è¤‡ã—ã¦ã„ã‚‹ã‚«ãƒ©ãƒ åã‚’ãƒ­ã‚°ã«å‡ºåŠ›ã—ã€ãƒ‡ãƒ¼ã‚¿è¨­è¨ˆã«ç•°å¸¸ãŒãªã„ã‹ãƒãƒªãƒ‡ãƒ¼ã‚·ãƒ§ãƒ³
print("duplicated columns:", dup_cols)
```

```python
# å„ã‚«ãƒ©ãƒ ã®æ¬ ææ•°ï¼ˆNull Countï¼‰ã‚’é›†è¨ˆã—ã€æ¬ æãŒå¤šã„é †ï¼ˆé™é †ï¼‰ã«ã‚½ãƒ¼ãƒˆã—ã¦è¦æ¨¡ã‚’æŠŠæ¡
na_count = df.isna().sum().sort_values(ascending=False)

# å…¨ãƒ‡ãƒ¼ã‚¿ã«å¯¾ã™ã‚‹æ¬ æç‡ï¼ˆNull Rateï¼‰ã‚’ç®—å‡ºã—ã€ãƒ‡ãƒ¼ã‚¿ã®ä¿¡é ¼åº¦ã‚’å®šé‡çš„ã«è©•ä¾¡
na_rate = (df.isna().mean() * 100).sort_values(ascending=False)

# æ¬ æã®å®Ÿæ•°ã¨å‰²åˆã‚’çµ±åˆã—ãŸè¦ç´„ãƒ†ãƒ¼ãƒ–ãƒ«ã‚’ä½œæˆã—ã€ä¸Šä½30ä»¶ã‚’ãƒ¬ãƒ³ãƒ€ãƒªãƒ³ã‚°ï¼ˆãƒ—ãƒ­ãƒ•ã‚¡ã‚¤ãƒªãƒ³ã‚°è¡¨ç¤ºï¼‰
display(pd.DataFrame({"na_count": na_count, "na_rate_%": na_rate}).head(30))
```

```python
# æ¬ ææ•°ã¨æ¬ æç‡ã®Seriesã‚’çµåˆã—ã€å“è³ªè©•ä¾¡ç”¨ã®ã‚µãƒãƒªãƒ†ãƒ¼ãƒ–ãƒ«ã‚’ä½œæˆ
df_na_summary = pd.DataFrame({"na_count": na_count, "na_rate_%": na_rate})

# æ¬ æã®ãªã„æ­£å¸¸ãªåˆ—ã‚’é™¤å¤–ã—ã€å¯¾å¿œãŒå¿…è¦ãªã‚«ãƒ©ãƒ ã®ã¿ã‚’ã‚¯ã‚¨ãƒªæŠ½å‡ºã—ã¦è¡¨ç¤º
display(df_na_summary.query("na_count > 0"))
```

```python
# å…¨ã‚«ãƒ©ãƒ ã®å€¤ãŒå®Œå…¨ã«ä¸€è‡´ã™ã‚‹è¡Œï¼ˆå®Œå…¨é‡è¤‡ãƒ¬ã‚³ãƒ¼ãƒ‰ï¼‰ã‚’æ¤œçŸ¥ã—ã€ãã®ç·æ•°ã‚’ã‚«ã‚¦ãƒ³ãƒˆ
print("duplicate rows:", df.duplicated().sum())
```

```python
# æ¤œè¨¼å¯¾è±¡ã¨ã™ã‚‹ãƒ¦ãƒ‹ãƒ¼ã‚¯ã‚­ãƒ¼å€™è£œï¼ˆè­˜åˆ¥å­ï¼‰ã‚’å¤‰æ•°å®šç¾©
key = "prefecture"

# å¯¾è±¡ã‚­ãƒ¼ã«ãŠã‘ã‚‹ä¸€æ„æ€§åˆ¶ç´„ï¼ˆUniquenessï¼‰é•åã®ç·æ•°ã‚’æ¤œçŸ¥ã—ã¦ãƒ­ã‚°å‡ºåŠ›
print("duplicated key:", df[key].duplicated().sum())

# é‡è¤‡ãƒ¬ã‚³ãƒ¼ãƒ‰ã‚’å…¨ä»¶ï¼ˆkeep=Falseï¼‰æŠ½å‡ºã—ã€å†…å®¹ç…§åˆã®ãŸã‚ã«ã‚½ãƒ¼ãƒˆã—ã¦è¡¨ç¤º
df[df[key].duplicated(keep=False)].sort_values(key)
```

```python
# æ•°å€¤å‹ã®ç‰¹å¾´é‡ï¼ˆé‡çš„å¤‰æ•°ï¼‰ã®ã¿ã‚’æŠ½å‡ºã—ã€çµ±è¨ˆå‡¦ç†ç”¨ã®ãƒªã‚¹ãƒˆã¨ã—ã¦å®šç¾©
num_cols = df.select_dtypes(include=[np.number]).columns.tolist()

# æ•°å€¤ä»¥å¤–ã®ã‚«ãƒ†ã‚´ãƒªå¤‰æ•°ï¼ˆè³ªçš„å¤‰æ•°ï¼‰ã‚’åˆ†é›¢ã—ã€ã‚¨ãƒ³ã‚³ãƒ¼ãƒ‡ã‚£ãƒ³ã‚°ç­‰ã®å€‹åˆ¥ãƒãƒ³ãƒ‰ãƒªãƒ³ã‚°ç”¨ã«ãƒªã‚¹ãƒˆåŒ–
obj_cols = df.select_dtypes(exclude=[np.number]).columns.tolist()

# ç‰¹å¾´é‡ã®å†…è¨³ï¼ˆæ¬¡å…ƒæ•°ï¼‰ã‚’ãƒ­ã‚°å‡ºåŠ›ã—ã€ãƒ‡ãƒ¼ã‚¿ã®æ§‹æˆæ¯”ç‡ã‚’æŠŠæ¡
print("numeric:", len(num_cols))
print("non-numeric:", len(obj_cols))

# ã‚«ãƒ†ã‚´ãƒªå¤‰æ•°ã®ã‚«ãƒ©ãƒ åï¼ˆã‚¹ã‚­ãƒ¼ãƒï¼‰ã‚’å‡ºåŠ›ã—ã€å¾Œç¶šã®å‡¦ç†æ–¹é‡ã‚’ç¢ºèª
print("non-numeric cols:", obj_cols)
```

```python
# æ•°å€¤ãƒ‡ãƒ¼ã‚¿ã®åŸºæœ¬çµ±è¨ˆé‡ã‚’ç®—å‡ºã—ã€ç‰¹å¾´é‡ãŒå¤šã„å ´æ‰€ã§ã‚‚è¦–èªã—ã‚„ã™ã„ã‚ˆã†è»¢ç½®(Transpose)
stats = df[num_cols].describe().T

# å¹³å‡å€¤(mean)ã¨ã®ä¹–é›¢(åˆ†å¸ƒã®æ­ªã¿)ã‚’æ¯”è¼ƒæ¤œè¨¼ã™ã‚‹ãŸã‚ã€ä¸­å¤®å€¤ã‚’æ˜ç¤ºçš„ã«åˆ—è¿½åŠ 
stats["median"] = df[num_cols].median()

# ç”Ÿæˆã—ãŸçµ±è¨ˆãƒ—ãƒ­ãƒ•ã‚¡ã‚¤ãƒ«ã‚’ãƒ¬ãƒ³ãƒ€ãƒªãƒ³ã‚°ã—ã€å¤–ã‚Œå€¤ã‚„ç•°å¸¸æ¤œçŸ¥ã®ä¸€æ¬¡åˆ¤æ–­ã‚’è¡Œã†
display(stats)
```

```python
# æ•°å€¤åˆ—ã‚’å…¨ä»¶ã‚¹ã‚­ãƒ£ãƒ³ã—ã€è«–ç†çš„ã«ä¸æ­£ãªå¯èƒ½æ€§ãŒã‚ã‚‹ã€Œè² ã®å€¤ã€ã‚’å«ã‚€ã‚«ãƒ©ãƒ ã‚’ãƒªã‚¹ãƒˆåŒ–
neg_cols = [c for c in num_cols if (df[c]< 0).any()]

# ç•°å¸¸æ¤œçŸ¥ï¼ˆAnomaly Detectionï¼‰ã®çµæœã¨ã—ã¦ã€ãƒã‚¤ãƒŠã‚¹å€¤ã‚’æŒã¤ã‚«ãƒ©ãƒ åã‚’ãƒ­ã‚°å‡ºåŠ›
print("has negative values",neg_cols)
```

```python
# ã‚«ãƒ¼ãƒ‡ã‚£ãƒŠãƒªãƒ†ã‚£ï¼ˆãƒ¦ãƒ‹ãƒ¼ã‚¯æ•°ï¼‰ãŒ1ä»¥ä¸‹ã®ã€Œåˆ†æ•£0ã€ã®ç‰¹å¾´é‡ã‚’ç‰¹å®šã—ã€æƒ…å ±é‡ã®ãªã„ã‚«ãƒ©ãƒ ã‚’æŠ½å‡º
zero_var = [c for c in num_cols if df[c].nunique(dropna=True) <= 1]

# åˆ†æãƒ¢ãƒ‡ãƒ«ã«å¯„ä¸ã—ãªã„ã€Œå‰Šé™¤å€™è£œï¼ˆPruning Candidatesï¼‰ã€ã¨ã—ã¦ãƒ­ã‚°å‡ºåŠ›
print("zero/constant columns:", zero_var)
```

```python
def iqr_outliers(s: pd.Series):
    # æ¬ æå€¤ï¼ˆNaNï¼‰ã¯çµ±è¨ˆè¨ˆç®—ã®é‚ªé­”ã«ãªã‚‹ãŸã‚ã€å‰å‡¦ç†ã¨ã—ã¦é™¤å¤–
    s = s.dropna()

    # ãƒ‡ãƒ¼ã‚¿ã®ç¨®é¡ï¼ˆã‚«ãƒ¼ãƒ‡ã‚£ãƒŠãƒªãƒ†ã‚£ï¼‰ãŒå°‘ãªã™ãã‚‹å ´åˆã¯ã€çµ±è¨ˆçš„ãªåˆ¤å®šãŒä¸èƒ½ãªãŸã‚ã‚¹ã‚­ãƒƒãƒ—
    if s.nunique() < 5:
        return pd.DataFrame()

    # ãƒ‡ãƒ¼ã‚¿ã®åˆ†å¸ƒã«ãŠã‘ã‚‹ç¬¬1å››åˆ†ä½æ•°ï¼ˆ25%ç‚¹ï¼‰ã¨ç¬¬3å››åˆ†ä½æ•°ï¼ˆ75%ç‚¹ï¼‰ã‚’ç®—å‡º
    q1, q3 = s.quantile([0.25, 0.75])

    # ãƒ‡ãƒ¼ã‚¿ã®ä¸­å¤®50%ãŒå«ã¾ã‚Œã‚‹ç¯„å›²ï¼ˆIQRï¼šå››åˆ†ä½ç¯„å›²ï¼‰ã‚’è¨ˆç®—ã—ã€ã°ã‚‰ã¤ãã‚’å®šé‡åŒ–
    iqr = q3 - q1

    # å¤–ã‚Œå€¤ã¨ã¿ãªã™ä¸‹é™ãƒ»ä¸Šé™ã®é–¾å€¤ï¼ˆFenceï¼‰ã‚’å®šç¾©ï¼ˆä¸€èˆ¬çš„ãª1.5å€ãƒ«ãƒ¼ãƒ«ã‚’é©ç”¨ï¼‰
    lo, hi = q1 - 1.5*iqr, q3 + 1.5*iqr

    # è¨ˆç®—ã—ãŸé–¾å€¤ã¨ã€å®Ÿéš›ã®ãƒ‡ãƒ¼ã‚¿ã®æœ€å°ãƒ»æœ€å¤§å€¤ã‚’æ§‹é€ åŒ–ãƒ‡ãƒ¼ã‚¿ã¨ã—ã¦è¿”å´
    return pd.DataFrame({"low": [lo], "high": [hi], "min": [s.min()], "max": [s.max()]})

# çµæœã‚’æ ¼ç´ã™ã‚‹ç©ºã®ãƒªã‚¹ãƒˆï¼ˆã‚³ãƒ³ãƒ†ãƒŠï¼‰ã‚’åˆæœŸåŒ–
out_summary = []

# å…¨ã¦ã®æ•°å€¤ã‚«ãƒ©ãƒ ã«å¯¾ã—ã¦ã‚¤ãƒ†ãƒ¬ãƒ¼ã‚·ãƒ§ãƒ³ï¼ˆåå¾©å‡¦ç†ï¼‰ã‚’å®Ÿè¡Œ
for c in num_cols:
    # å®šç¾©ã—ãŸé–¢æ•°ã‚’å‘¼ã³å‡ºã—ã€å¤–ã‚Œå€¤åˆ¤å®šã‚’å®Ÿè¡Œ
    r = iqr_outliers(df[c])

    # çµæœãŒç©ºã§ãªã„ï¼ˆåˆ¤å®šã§ããŸï¼‰å ´åˆã®ã¿å‡¦ç†
    if not r.empty:
        # ã©ã®ã‚«ãƒ©ãƒ ã®çµæœã‹åˆ†ã‹ã‚‹ã‚ˆã†ã«ã€è­˜åˆ¥ç”¨ã®åˆ—ã‚’å…ˆé ­ã«æŒ¿å…¥
        r.insert(0, "col", c)
        out_summary.append(r)

# çµæœãŒå­˜åœ¨ã™ã‚‹å ´åˆã®ã¿ã€ãƒªã‚¹ãƒˆã‚’çµåˆï¼ˆConcatï¼‰ã—ã¦ãƒ¬ãƒãƒ¼ãƒˆã‚’è¡¨ç¤º
if out_summary:
    display(pd.concat(out_summary, ignore_index=True).sort_values("col"))
```

```python
# åˆ†æã®ã€Œç›®çš„å¤‰æ•°ï¼ˆã‚¿ãƒ¼ã‚²ãƒƒãƒˆï¼‰ã€ã¨ãªã‚‹ã€æœ€é‡è¦ã®é›¢è·ç‡é–¢é€£ã‚«ãƒ©ãƒ ã‚’å®šç¾©
turnover_cols = ["turnover_total", "turnover_new_grad", "turnover_experienced"]

# ç›®çš„å¤‰æ•°ã®ã„ãšã‚Œã‹ã«æ¬ æï¼ˆNaNï¼‰ãŒã‚ã‚‹è¡Œã‚’ç‰¹å®šã—ã€åˆ†æã«ä½¿ãˆãªã„ã€Œä¸å®Œå…¨ãƒ‡ãƒ¼ã‚¿ã€ã¨ã—ã¦æŠ½å‡º
# axis=1 ã¯ã€Œè¡Œå˜ä½ï¼ˆæ¨ªæ–¹å‘ï¼‰ã€ã«ãƒã‚§ãƒƒã‚¯ã™ã‚‹ã¨ã„ã†æ„å‘³
missing_turnover = df[df[turnover_cols].isna().any(axis=1)]

# æ¬ æã«ã‚ˆã‚‹ã€Œãƒ‡ãƒ¼ã‚¿ãƒ­ã‚¹ï¼ˆåˆ†æå¯¾è±¡å¤–ã¨ãªã‚‹ä»¶æ•°ï¼‰ã€ã‚’ãƒ­ã‚°å‡ºåŠ›ã—ã€å½±éŸ¿ã®å¤§ãã•ã‚’æ¸¬å®š
print("rows with missing turnover:", len(missing_turnover))

# æ¬ æã®ç™ºç”Ÿå‚¾å‘ï¼ˆç‰¹å®šã®éƒ½é“åºœçœŒã ã‘ã§èµ·ãã¦ã„ã‚‹ã‹ï¼Ÿãªã©ï¼‰ã‚’ç›®è¦–ç¢ºèªã™ã‚‹ãŸã‚ã€å±æ€§ã‚’çµã£ã¦ã‚µãƒ³ãƒ—ãƒ«è¡¨ç¤º
display(missing_turnover[["prefecture"] + turnover_cols].head(20))
```

```python
# ã‚«ãƒ†ã‚´ãƒªå¤‰æ•°ï¼ˆè³ªçš„å¤‰æ•°ï¼‰ã®å…¨ã‚«ãƒ©ãƒ ã«å¯¾ã—ã¦ã€ãƒ‡ãƒ¼ã‚¿ã®ä¸­èº«ã‚’ç¢ºèªã™ã‚‹ã‚¤ãƒ†ãƒ¬ãƒ¼ã‚·ãƒ§ãƒ³ï¼ˆåå¾©å‡¦ç†ï¼‰ã‚’é–‹å§‹
for c in obj_cols:
    # æ¬ æã‚’é™¤å»ã—ã¦æ–‡å­—åˆ—å‹ã«ã‚­ãƒ£ã‚¹ãƒˆï¼ˆå‹å¤‰æ›ï¼‰ã—ã€å®Ÿãƒ‡ãƒ¼ã‚¿ã‚’å…ˆé ­5ä»¶ã ã‘ã‚µãƒ³ãƒ—ãƒªãƒ³ã‚°ï¼ˆæŠ½å‡ºï¼‰
    sample = df[c].dropna().astype(str).head(5).tolist()
    
    # ãƒ‡ãƒ¼ã‚¿ã®ã€Œãƒ•ã‚©ãƒ¼ãƒãƒƒãƒˆã€ã‚„ã€Œå…¥åŠ›è¦å‰‡ã€ã‚’ç›®è¦–ãƒã‚§ãƒƒã‚¯ï¼ˆã‚¤ãƒ³ã‚¹ãƒšã‚¯ã‚·ãƒ§ãƒ³ï¼‰ã™ã‚‹ãŸã‚ã€ã‚µãƒ³ãƒ—ãƒ«å€¤ã‚’ãƒ­ã‚°å‡ºåŠ›
    print(c,"sample",sample)
```
