# 05. 解釈と提言（Interpretation & Policy Implications）

## 📌 Key Insights（結論）
本プロジェクトの回帰モデル（Model_D）は、都道府県別の看護師離職率のばらつきを **高い精度で説明** し、離職が「個人の問題」ではなく **構造要因（労働負荷 × 市場環境 × 生活コスト）** の組み合わせで起きている可能性を示唆する。

- **Workload（夜勤負荷）**：夜勤過多は離職を押し上げる方向
- **Labor Market（求人環境）**：求人倍率が高い地域では離職が下がる方向（競争圧による待遇改善の可能性）
- **Cost of Living（家賃）**：生活コストの高さは離職を押し上げる方向

> **⚠️ Interpretation Note（解釈上の注意）**  
> 本Notebookで述べる内容は **相関にもとづく解釈** であり、因果関係を直接証明するものではない（集計データ分析の制約／生態学的誤謬のリスク）。

| 要因 | 係数 (Coef) | **現実スケール換算（Impact）** | 方向 |

※ 係数の解釈は **他の変数を一定としたときの平均的な関係** を意味する。


## 🧠 Structural Interpretation（要因別の構造解釈）

### ① 夜勤負担（night_shift_72h_plus）
- **示唆**：夜勤過多が常態化している地域ほど離職が高い方向。  
- **現場スケール**：たとえば「夜勤72時間超割合」が数％動くだけで、離職率の差としては無視できない幅になり得る。  
- **解釈（仮説）**：長時間・高頻度の夜勤は、疲労蓄積 → 回復不能 → バーンアウトの経路で離職を押し上げる可能性がある。

### ② 有効求人倍率（job_openings_ratio）
- **示唆**：求人倍率が高い地域ほど離職が下がる方向（逆説的に見える）。  
- **解釈（仮説）**：  
  - 競争圧が働き、待遇改善（賃金・働き方・教育体制）が進む  
  - 「辞めてもすぐ次がある」環境でも、そもそも辞める必要がない状態が作られている  
  など、**市場構造が定着に効いている** 可能性。

### ③ 家賃相場（rent_private）
- **示唆**：家賃が高い地域ほど離職が高い方向。  
- **解釈（仮説）**：生活コストの圧迫は、可処分所得や生活満足度を下げ、結果として定着を弱める可能性がある。  
- **注意**：家賃は都市性（人口集中、物価、病院規模）などの代理変数にもなり得るため、解釈は慎重に扱う。


## 🏥 Policy Implications（施策提言）
本分析（全体モデル）の示唆に基づき、離職抑制に向けた **3つの構造的アプローチ** を整理する。  
※ ここでの提言は「看護師全体」の傾向にもとづく。キャリアステージ（新卒／既卒）の違いは `06_subgroup_analysis.ipynb` で追加検証する。

### 提言A：夜勤負担の「量的削減」と「質的転換」
- **量（Workload）**：夜勤の回数・連続性・回復時間の設計を見直す  
- **質（Work Design）**：夜勤帯の業務構造（巡視・記録・対応フロー）を見直す  
- **具体例**  
  - タスク・シフティング（看護補助者／他職種への移譲）  
  - 記録・見守りの省力化（音声入力、センサー等）  
  - 夜勤配置の再設計（属人化の解消、標準化）

### 提言B：都市部と地方の「採用・定着戦略」の差別化
- **都市部**：生活コスト（住居）への支援設計が効きやすい  
- **地方**：人材流出の構造（キャリア形成、教育、家族要因）への対策が効きやすい  
- **具体例**  
  - 住宅手当・寮・補助制度の設計（都市部）  
  - 教育・キャリアラダー・専門性支援（地方含む）

### 提言C：市場環境を「味方につける」定着施策
求人倍率の高さが離職抑制と関連しているなら、競争環境を前提に **“辞めない理由”** を作る設計が重要になる。
- 賃金だけでなく、働き方・教育・評価制度を含めた総合設計  
- 採用強化と同時に、定着のKPI（夜勤負荷、教育完了率、休暇取得など）を監視する仕組みづくり


## ⚠ Limitations & Future Work（限界と次の問い）

### Limitations（本分析の限界）
- **因果は示さない**：相関にもとづく解釈であり、介入効果を直接推定するものではない  
- **集計データの制約**：都道府県単位の平均であり、個人・病院レベルの異質性を潰している  
- **サンプルサイズ**：47都道府県のため、モデルの安定性・外れ値影響に限界がある  
- **未観測要因**：病院規模、病床機能、教育体制、離職の定義差などの影響が残り得る
