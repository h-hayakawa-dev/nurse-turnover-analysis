# 04. 解釈と提言（Interpretation & Policy Implications）

## 0. 分析設定（Analysis Context）
本レポートの結論は、前段のモデリングプロセス（`03_modeling.ipynb`）で選定された以下の最終モデルに基づいている。

* **データソース:** 47都道府県 公的統計データ（2023年時点）<br> ※データの年次について 分析の基準年は「2023年度」としているが、一部の指標については、利用可能な直近データを使用している。
* **詳細:** データの出所やクリーニング処理については `01_data_validation.ipynb` を参照
* **最終採用モデル:** **Model_D**（決定係数 $R^2_{adj} = 0.81$）
* **目的変数:** 常勤看護師の離職率（Turnover Rate）
* **説明変数:** ①夜勤72時間超割合、②有効求人倍率、③家賃相場（※持ち家率は有意でないため除外）

## 1. Executive Summary（分析結果の要約）

### 結論
看護師の離職率は、個人の資質よりも**「過重労働（夜勤）」と「市場環境（求人・生活コスト）」という3つの構造的要因**によって80%以上説明できる。

### 主要な発見（Key Findings）
1.  **夜勤負担が最大の離職要因:** 「72時間ルール」を超過する勤務が常態化している地域ほど、離職率が有意に高い。
2.  **求人が多いほど離職は減る（逆説）:** 有効求人倍率が高い（売り手市場の）地域ほど離職率は低い。これは競争原理による待遇改善が機能していることを示唆する。
3.  **生活コストの圧迫:** 家賃相場が高い地域では、経済的ストレスが離職を後押ししている。

> **⚠️ 解釈上の注意（Limitations）**
> * 本分析は都道府県単位の集計データを用いたものであり、個々の看護師や病院の因果関係を直接証明するものではない（生態学的誤謬のリスク）。
> * 提示する関係性は「統計的な相関」であり、解釈には現場の定性的背景（Field View）を補う必要がある。

## 2. 要因別の詳細解釈と「現実スケール」への翻訳

モデルの係数（Coefficient）を、現場で直感的に理解できる単位（1万円、1%、0.1倍）に換算して評価する。

| 要因 | 係数 (Coef) | **現実スケール換算（Impact）** | 解釈の方向 |
| :--- | :--- | :--- | :--- |
| **① 夜勤72時間超** | **+0.098** | 割合が **1%** 増えると、離職率は **約0.1ポイント** 上昇 | **促進** (Burnout) |
| **② 有効求人倍率** | **-4.477** | 倍率が **0.1** 上がると、離職率は **約0.45ポイント** 低下 | **抑制** (Retention) |
| **③ 家賃相場** | **+0.0001** | 家賃が **1万円** 上がると、離職率は **約1.0ポイント** 上昇 | **促進** (Cost Stress) |

### ① 夜勤負担（night_shift_72h_plus）
* **現場感覚への翻訳:**
    * 「たった0.1ポイント？」と思うかもしれないが、離職率の平均が10%前後であることを考えると、夜勤負担の数%の違いが**離職率を1ポイント（10%相当）押し上げる**影響力は甚大である。
    * **Field View:** これは単なる長時間労働ではなく、不規則勤務による「累積疲労」と、余裕のない職場の「荒廃した空気」を反映している。

### ② 有効求人倍率（job_openings_ratio）
* **現場感覚への翻訳:**
    * 求人倍率が0.1改善する（例: 1.2倍→1.3倍）だけで、離職率が0.5ポイント近く下がる。市場の「引く手あまた感」は、むしろ病院側の**「待遇改善（給与・福利厚生）への本気度」**を引き出し、結果として離職を食い止めている。
    * **Field View:** 逆に求人がない地域では、「辞められない」足元を見た労務管理がまかり通っているリスクがある。

### ③ 家賃（rent_private）
* **現場感覚への翻訳:**
    * 家賃相場が1万円違うだけで、離職率に約1ポイントの差が出る。
    * **Field View:** 給与が変わらないのに家賃だけ高い（可処分所得が低い）状態は、若手ナースにとって「生活できない」という切実な退職理由に直結する。

## 3. 施策提言（Policy Implications）

本分析（全体モデル）の結果に基づき、看護師の離職抑制に向けた3つの構造的アプローチを提言する。
※ ただし、これらは「看護師全体」の傾向に基づくものであり、キャリアステージ（新卒・既卒）による違いは後述のサブグループ分析で検証する。

### 提言A：夜勤負担の「量的削減」と「質的転換」
* **背景:**
    * 夜勤72時間超えが常態化している地域ほど離職率が高い。これは個人の努力で解決できる問題ではない。
* **具体的アクション:**
    * **タスク・シフティングの加速:**
        * 看護補助者（ナースエイド）への業務移譲を徹底し、看護師しかできない業務に集中させることで、夜勤人員の適正配置を行う。
    * **テクノロジーによる業務効率化:**
        * AI音声入力による記録時間の短縮や、見守りセンサー導入により、夜勤中の「巡視・記録」という二大負荷を物理的に削減する。

### 提言B：都市部と地方の「採用・定着戦略」の差別化
* **背景:**
    * 市場環境（求人倍率・家賃）によって離職メカニズムが異なる可能性がある。
* **具体的アクション:**
    * **都市部（高家賃・高求人倍率エリア）:**
        * 転職の選択肢が豊富なため、「給与」だけでなく「生活支援（家賃補助・寮）」や「ワークライフバランス」を差別化要因とする。
        * 「辞めなくてもここで働き続けたい」と思わせるエンゲージメント施策が最優先。
    * **地方部（低家賃・低求人倍率エリア）:**
        * 「他に働く場所がない」という消極的定着に甘んじず、労働環境（特に夜勤負担）を改善することで、地域内でのブランド力を高める。
        * 潜在看護師の掘り起こしを行い、マンパワー不足による夜勤負担増の悪循環を断ち切る。

## 4. モデルの限界と次なる問い（Limitations & Future Work）

### 本分析の限界
今回の分析は「看護師全体」の平均値を用いたマクロ分析であり、キャリアステージによる「離職理由の質的な違い」までは見えていない。

### 次なる問い（Next Step）
現場感覚に基づくと、**「新卒（New Grads）」と「既卒・中堅（Existing Grads）」では、離職のメカニズムが全く異なる**という仮説が立てられる。

#### 仮説：キャリアステージによる「痛み」の違い

| 対象 | 仮説される主な離職トリガー | メカニズム（現場の肌感覚） |
| :--- | :--- | :--- |
| **新卒** | **環境・人間関係**<br>（夜勤負担の間接的影響） | 若さゆえに夜勤の体力負担そのものには耐えられるかもしれない。しかし、過重労働で余裕を失った**「先輩のピリピリした空気」や「不十分な教育環境」**に耐えられず、早期離職に至るのではないか。 |
| **既卒** | **待遇・納得感**<br>（給与 vs 負担） | ある程度業務には慣れているが、**「負担が増えても給料が上がらない」**という現実に直面し、労働対価への不満（相対的剥奪感）から離職を選ぶのではないか。 |
