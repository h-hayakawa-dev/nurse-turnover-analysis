# 00. データ前処理（Preprocessing / ETL）

## 目的
- raw（元データ）を、分析しやすい表形式に整形して processed に保存する
- Excelの「印刷用レイアウト（タイトル行・空行・結合セル）」をデータ形式に変換する

## 出力
- data/processed/ に加工済みCSVを出力する


② Codeセル（読み込み：header調整）

まずは ファイル名だけあなたのものに変えてください。

```python
import pandas as pd
from pathlib import Path

raw_path = Path("../data/raw/令和５年住宅・土地統計調査　住宅及び世帯に関する基本集計.xlsx")
raw_path
```

```python
df = pd.read_excel(raw_path, header=3)
df.head(10)
```

```python
df = pd.read_excel(raw_path, header=4)
df.head(10)
```

```python
df = df.dropna(axis=1, how="all")  # 全部NaNの列を削除
df.columns
```

```python
df = df.dropna(how="all")  # 全部NaNの行を削除
df.head(10)
```

```python
out_path = Path("../data/processed/housing_land_survey_2023_clean.csv")
out_path.parent.mkdir(parents=True, exist_ok=True)

df.to_csv(out_path, index=False)
print("saved:", out_path)
```
