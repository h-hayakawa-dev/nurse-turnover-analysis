# 04. 解釈と提言（Interpretation & Policy Implications）

## 1. Executive Summary（分析結果の要約）

本プロジェクトでは、都道府県別の公的統計データを用い、看護師離職率（Turnover Rate）を説明する構造的要因を分析した。
検証の結果、個人の資質やモチベーション論ではなく、以下の**3つの外部環境要因**が離職率の変動を80%以上説明することが明らかになった（Model D, Adj R2 = 0.81）。

### 主要な発見（Key Findings）

1.  **夜勤負担が最大の離職要因である**
    * 「夜勤72時間超えの看護師割合」が高い地域ほど、明確に離職率が高い。これは過重労働が燃え尽き（Burnout）を引き起こす直接的な構造要因であることを示唆している。

2.  **「求人が多い地域」ほど、逆に離職率は低い（逆説的発見）**
    * 有効求人倍率（job_openings_ratio）が高い都市部エリアほど、離職率が低いという負の相関（係数: -4.48）が確認された。
    * これは「売り手市場（求人が多い）であるため、病院側が人材確保のために労働条件や待遇を改善せざるを得ない」という**市場の競争原理（Market Mechanism）**が機能している結果と解釈できる。

3.  **生活コスト（家賃）が隠れたストレス要因**
    * 家賃相場（rent_private）が高い地域では離職率が高まる傾向にある。給与水準が生活コストの上昇に追いついていない場合、経済的なストレスが離職を後押ししている可能性がある。

## 2. 要因別の詳細解釈（Factor Analysis & Field View）

最終モデル（Model D）の各係数について、データが示す事実と、**臨床現場の視点（Field View）** を融合して解釈する。

| 変数名 | 日本語ラベル | 係数 (Coef) | P値 | 解釈の方向性 |
| :--- | :--- | :--- | :--- | :--- |
| **night_shift_72h_plus** | 夜勤72時間超割合 | **+0.098** | <0.01 | **促進要因**（増えると離職が増える） |
| **job_openings_ratio** | 有効求人倍率 | **-4.477** | <0.01 | **抑制要因**（増えると離職が減る） |
| **rent_private** | 家賃（1畳あたり） | **+0.0001** | <0.01 | **促進要因**（増えると離職が増える） |

### ① 夜勤負担（night_shift_72h_plus）
* **データ解釈:**
    * 係数はプラス（+0.098）。「72時間ルール」を守れない施設が多い地域ほど、離職率は有意に高い。
* **現場視点（Field View）:** **「累積的疲労と職場の荒廃」**
    * 夜勤負担は「その瞬間に辞める」トリガーではなく、「散り積もるしんどさ」として蓄積する。月100時間を超えるような過重労働は、睡眠リズムを恒久的に破壊し、回復不可能な健康被害をもたらす。
    * また、人手不足で余裕がない現場特有の**「ピリピリした空気」**を醸成し、間接的に人間関係を悪化させる要因ともなっている。

### ② 有効求人倍率（job_openings_ratio）
* **データ解釈:**
    * 係数はマイナス（-4.48）。求人が多い（転職しやすい）地域ほど、逆に離職率が低い。
* **現場視点（Field View）:** **「経済的ロックインと危機感」**
    * 都市部は求人も多いが、家賃や生活コストも高いため、共働きであっても「今の給与水準」を手放せず、辞めたくても辞められない（Economic Lock-in）可能性がある。
    * また、競合が激しいエリアでは病院側も「人材流出の危機感」を持っており、地方に比べてリテンション対策（待遇改善）が進んでいる側面がある。

### ③ 家賃（rent_private）
* **データ解釈:**
    * 係数はプラス。都市化の影響を除いても、生活コスト（固定費）の高さは離職を押し上げる。
    * 若手看護師にとって、高い家賃負担は可処分所得を圧迫し、QOL（生活の質）を低下させる隠れたストレッサーとなっている。

## 3. 施策提言（Policy Implications）

本分析（全体モデル）の結果に基づき、看護師の離職抑制に向けた3つの構造的アプローチを提言する。
※ ただし、これらは「看護師全体」の傾向に基づくものであり、キャリアステージ（新卒・既卒）による違いは後述のサブグループ分析で検証する。

### 提言A：夜勤負担の「量的削減」と「質的転換」
* **背景:**
    * 夜勤72時間超えが常態化している地域ほど離職率が高い。これは個人の努力で解決できる問題ではない。
* **具体的アクション:**
    * **タスク・シフティングの加速:**
        * 看護補助者（ナースエイド）への業務移譲を徹底し、看護師しかできない業務に集中させることで、夜勤人員の適正配置を行う。
    * **テクノロジーによる業務効率化:**
        * AI音声入力による記録時間の短縮や、見守りセンサー導入により、夜勤中の「巡視・記録」という二大負荷を物理的に削減する。

### 提言B：都市部と地方の「採用・定着戦略」の差別化
* **背景:**
    * 市場環境（求人倍率・家賃）によって離職メカニズムが異なる可能性がある。
* **具体的アクション:**
    * **都市部（高家賃・高求人倍率エリア）:**
        * 転職の選択肢が豊富なため、「給与」だけでなく「生活支援（家賃補助・寮）」や「ワークライフバランス」を差別化要因とする。
        * 「辞めなくてもここで働き続けたい」と思わせるエンゲージメント施策が最優先。
    * **地方部（低家賃・低求人倍率エリア）:**
        * 「他に働く場所がない」という消極的定着に甘んじず、労働環境（特に夜勤負担）を改善することで、地域内でのブランド力を高める。
        * 潜在看護師の掘り起こしを行い、マンパワー不足による夜勤負担増の悪循環を断ち切る。

## 4. モデルの限界と次なる問い（Limitations & Future Work）

### 本分析の限界
今回の分析は「看護師全体」の平均値を用いたマクロ分析であり、キャリアステージによる「離職理由の質的な違い」までは見えていない。

### 次なる問い（Next Step）
現場感覚に基づくと、**「新卒（New Grads）」と「既卒・中堅（Existing Grads）」では、離職のメカニズムが全く異なる**という仮説が立てられる。

#### 仮説：キャリアステージによる「痛み」の違い

| 対象 | 仮説される主な離職トリガー | メカニズム（現場の肌感覚） |
| :--- | :--- | :--- |
| **新卒** | **環境・人間関係**<br>（夜勤負担の間接的影響） | 若さゆえに夜勤の体力負担そのものには耐えられるかもしれない。しかし、過重労働で余裕を失った**「先輩のピリピリした空気」や「不十分な教育環境」**に耐えられず、早期離職に至るのではないか。 |
| **既卒** | **待遇・納得感**<br>（給与 vs 負担） | ある程度業務には慣れているが、**「負担が増えても給料が上がらない」**という現実に直面し、労働対価への不満（相対的剥奪感）から離職を選ぶのではないか。 |

次章（05_subgroup_analysis）では、この仮説を検証するために、データを「新卒」と「既卒」に分割し、それぞれの離職構造の違いを明らかにする。
